```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# 
fig, ax = plt.subplots(figsize=(10, 8))

# 
n_points = 100
x = np.random.randn(n_points)
y = np.random.randn(n_points)
colors = np.random.rand(n_points)
scat = ax.scatter(x, y, c=colors, s=100, alpha=0.7, cmap='viridis')

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_title('dynamic scatter plot')
```

```python
def update(frame):
    x_new = x + np.random.randn(n_points) * 0.1
    y_new = y + np.random.randn(n_points) * 0.1
    colors_new = np.random.rand(n_points)

    # 
    scat.set_offsets(np.column_stack([x_new, y_new]))
    scat.set_array(colors_new)
    
    return scat,
```

```python
ani=animation.FuncAnimation(
    fig=fig,frames=1000,interval=50,blit=True,
    func=update
)
plt.show()
```
